```python
#url = 'http://www.atlaspiv.cz/?page=seznam-pivovaru'
breweries_urls = []    
    
def scraping_urls_of_breweries(url_link):
    code = GetCode(url_link)
    
    lenght_breweries_urls = len(code.findAll('li', {'style':'margin-bottom: 5px;'}))

    print(lenght_breweries_urls)

    for i in range(lenght_breweries_urls):
        d = code.findAll('li', {'style':'margin-bottom: 5px;'})[i].findChild().attrs['href']
        breweries_urls.append(d)

    print(breweries_urls)    
    len(breweries_urls)
    
```

```python
breweries_dictionary = []

def find_info_about_breweries(list_of_urls):

    for i in range(len(list_of_urls)):
        time.sleep(0.1)
        string_1 = str(list_of_urls[i])
        url = 'http://www.atlaspiv.cz/' + string_1
        brewery_url = GetCode(url)

        brewery_id = brewery_url.find('input', {'name':'rec_id'}).attrs['value']
        brewery_name = brewery_url.findAll('h1')[1].text
        brewery_category = brewery_url.findAll('h1')[1].next.next.next.next.text
        brewery_adress = brewery_url.find('table').find('td').next.next.next.text
        brewery_web = brewery_url.find('table').find('a',{'rel': 'nofollow'}).text

        count_of_beers = len(brewery_url.findAll('span',{'class': 'rating_stars'}))
        beer_menu = []
        beer_web = []

        for i in range(count_of_beers):
            e = brewery_url.findAll('span',{'class': 'rating_stars'})[i].find_previous('a').text
            beer_menu.append(e)
            f = brewery_url.findAll('span',{'class': 'rating_stars'})[i].find_previous('a').attrs['href']
            beer_web.append(f)


        g = {'brewery_id':brewery_id, 'name':brewery_name, 'category':brewery_category, 'adress':brewery_adress, 'web':brewery_web, 'count_of_beers':count_of_beers, 'beer_menu':beer_menu}

        breweries_dictionary.append(g)
```

```python
find_info_about_breweries(breweries_urls)
```
